```python
import requests
from bs4 import BeautifulSoup as BS
import re
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
URL = 'https://www.goodreads.com/shelf/show/booktok'

response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
type(soup)
```

## get book info from first page

```python
# https://stackoverflow.com/questions/5041008/how-to-find-elements-by-class
cards = soup.find_all("div", {"class": "elementList"})
```

```python
print(cards)
```

### get titles of first page

```python
#https://www.geeksforgeeks.org/difference-between-find-and-find_all-in-beautifulsoup-python/
#https://www.adamsmith.haus/python/answers/how-to-build-a-pandas-dataframe-with-a-for-loop-in-python
books = []
for card in cards:
    books.append(str(card.find('a', {'class': 'bookTitle'})))
```

```python
books
```

```python
#https://www.codegrepper.com/code-examples/html/regex+to+remove+html+tags+python
titles = [re.sub('<.*?>', '', x) for x in books]
titles
```

```python
#remove empty values from list
#https://www.geeksforgeeks.org/python-remove-empty-strings-from-list-of-strings/
while("None" in titles) :
    titles.remove("None")
    
titles
```

### get URLs of first page

```python
#https://stackoverflow.com/questions/3075550/how-can-i-get-href-links-from-html-using-python

#when just using href attr, returned duplicates, so had to add class attr
#https://pytutorial.com/beautifulsoup-attribute#beautifulsoup-find-all-by-multiple-attributes

hrefs = []
for link in soup.findAll('a', attrs={'class': 'bookTitle', 'href': re.compile("^/book/show/")}) :
    hrefs.append(link.get('href'))
    
hrefs
```

```python
#https://towardsdatascience.com/scraping-multiple-urls-with-python-tutorial-2b74432d085f
string = 'https://www.goodreads.com'
urls = [string + s for s in hrefs]
urls
```

### get number of times shelved as "booktok"

```python
shelved_booktok = []
for card in cards:
    shelved_booktok.append(str(card.find(string=re.compile("shelved"))))
    
shelved_booktok
```

```python
#remove all but number of times shelved as booktok
#https://stackoverflow.com/questions/1249388/removing-all-non-numeric-characters-from-string-in-python
shelved_booktok = [re.sub("[^0-9]", "", x) for x in shelved_booktok]
shelved_booktok
```

```python
#remove empty values from list
#https://www.geeksforgeeks.org/python-remove-empty-strings-from-list-of-strings/
while("" in shelved_booktok) :
    shelved_booktok.remove("")
    
shelved_booktok
```

### get authors

```python
authors = []
for card in cards:
    authors.append(str(card.find('span', {'itemprop': 'name'})))
    
authors
```

```python
#https://www.codegrepper.com/code-examples/html/regex+to+remove+html+tags+python
authors = [re.sub('<.*?>', '', x) for x in authors]
authors
```

```python
#remove none values
#https://www.geeksforgeeks.org/python-remove-empty-strings-from-list-of-strings/
while("None" in authors) :
    authors.remove("None")
    
authors
```

## dataframe of first page

```python
first_page = {'title':titles, 'author':authors, 'times_shelved_as_booktok':shelved_booktok, 'URL':urls}
first_page = pd.DataFrame(first_page)
first_page.head(20)
```

```python
#https://data36.com/scrape-multiple-web-pages-beautiful-soup-tutorial/
#https://www.geeksforgeeks.org/how-to-scrape-multiple-pages-of-a-website-using-python/

#initiate data storage
books = []
titles = []
authors = []
shelved_book = []
hrefs = []
urls = []

url = "https://www.goodreads.com/shelf/show/booktok?page="
for page in range(1,31):
      
    response = requests.get(url + str(page))
    html = response.text
    soup = BS(html, 'html.parser')
    cards = soup.find_all("div", {"class": "elementList"})

    for card in cards:
        books.append(str(card.find('a', {'class': 'bookTitle'})))
        authors.append(str(card.find('span', {'itemprop': 'name'})))
        shelved_booktok.append(str(card.find(string=re.compile("shelved"))))
        
    #page = page + 1
```
