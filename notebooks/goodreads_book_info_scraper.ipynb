```python
import requests
from bs4 import BeautifulSoup as BS
import re
import pandas as pd
```

```python
#read in csv of booktook bestsellers
books_df = pd.read_csv('../data/booktok_books.csv')
```

```python
books_df.head()
```

```python
#get list of urls from url column
url_list = books_df['url'].tolist()
print(url_list)
```

```python
#scrape book info from each url
avg_rating = []

for i in range(0,6):
    url = url_list[i]
    response = requests.get(url)
    soup = BS(response.text, 'html.parser')
    ratings = soup.find_all("div", {"id": "bookMeta"})

    #for tag in my_table2:
        #data.append(tag.text.strip())
        
    for r in ratings:
        avg_rating.append(str(ratings.find('span', {'itemprop': 'ratingValue'})))
        
avg_rating
```
