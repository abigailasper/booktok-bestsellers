```python
#read in csv of booktook bestsellers
books_df = pd.read_csv('../data/booktok_books.csv')
```

```python
books_df.head()
```

```python
#get list of urls from url column
url_list = books_df['url'].tolist()
print(url_list)
```

```python
print(len(url_list))
```

```python
#scrape book info from each url
avg_rating = []
count_ratings = []
num_pages = []
date_published = []
top_genre = []

for i in range(0,1251):
    url = url_list[i]
    driver.get(url)  
    sleep(randint(10,20))
    soup = BS(driver.page_source, 'html.parser')
    ratings = soup.find_all('span', {'itemprop': 'ratingValue'})
    num_ratings = soup.find_all('meta', {'itemprop': 'ratingCount'})
    details = soup.find_all('div', {'id': 'details'})
    genre = soup.find_all('a', {'class': 'actionLinkLite bookPageGenreLink'})[0]
    
    for r in ratings:
        avg_rating.append(r)
        
    for n in num_ratings:
        count_ratings.append(n)
        
    for i in details:
        num_pages.append(str(i.find(string=re.compile("pages"))))
        date_published.append(str(i.find(string=re.compile("Published"))))
            
    for g in genre:
        top_genre.append(g)
```

```python
#remove html tags
#https://www.codegrepper.com/code-examples/html/regex+to+remove+html+tags+python
count_ratings = str(count_ratings)
count_ratings = [re.sub('<.*?>', '', x) for x in count_ratings]
count_ratings
```

```python
#remove all but number of times shelved as booktok
#https://stackoverflow.com/questions/1249388/removing-all-non-numeric-characters-from-string-in-python
count_ratings = [re.sub("[^0-9]", "", x) for x in count_ratings]
count_ratings
```

```python
book_info1 = {'avg_rating':avg_rating, 'count_ratings':count_ratings, 'num_pages':num_pages, 'date_published':date_published, 'top_genre':top_genre}
book_info1 = pd.DataFrame(book_info1)
book_info1.head(20)
```
