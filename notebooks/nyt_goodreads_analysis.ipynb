# clean data

## clean nyt bestseller data

```python
#drop columns that aren't needed

bestsellers_df = bestsellers_df.drop(['Unnamed: 0', 'primary_isbn10', 'price', 'contributor', 'contributor_note', 'book_image_width', 'book_image_height', 'book_review_link', 'first_chapter_link', 'sunday_review_link', 'article_chapter_link', 'isbns', 'buy_links', 'book_uri', 'age_group'], axis=1)
bestsellers_df.head()
```

### create df of only most recent appearances on nyt bestseller list

```python
#https://www.geeksforgeeks.org/delete-duplicates-in-a-pandas-dataframe-based-on-two-columns/
# drop rows which have same title
# and author and keep latest entry
bestsellers_recent_df = bestsellers_df.drop_duplicates(
  subset = ['title', 'author'],
  keep = 'last').reset_index(drop = True)
```

## clean goodreads booktok data

```python
#make booktok_popularity rank start at 1 instead of 0
booktok_df['booktok_popularity'] = booktok_df['booktok_popularity'] + 1
booktok_df.head()
```

```python
#rename title column so that it can be expanded into two columns: title + series_format info
booktok_df.rename(columns = {"title": "title_and_specs"}, 
          inplace = True)
booktok_df.head()
```

```python
#expand title_and_specs into individual columns
#https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/
#https://stackoverflow.com/questions/42205616/using-str-split-for-pandas-dataframe-values-based-on-parentheses-location
#https://datascienceparichay.com/article/pandas-split-column-by-delimiter/
#https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html#pandas.Series.str.split
new_title = booktok_df['title_and_specs'].str.split("\s+\(", n=1, expand=True)
new_title.head()
```

```python
#change column names
new_title.columns = ['title', 'series_format']
```

```python
#drop closed parenthisis from series_format
new_title['series_format'] = new_title['series_format'].str.rstrip('\s+\)')
new_title.head()
```

```python
#make titles all caps to match nyt bestsellers titles
new_title['title'] = new_title['title'].str.upper()
new_title.head()
```

```python
#add new title and series_format columns back into booktok_df
booktok_df['title'] = new_title['title']
booktok_df['series_format'] = new_title['series_format']
booktok_df.head()
```

```python
#remove "pages" from num_pages column
booktok_df['num_pages'] = booktok_df['num_pages'].astype(str)
booktok_df['num_pages'] = booktok_df['num_pages'].str.rstrip(' pages')
booktok_df.head()
```
